# Grade Distribution analysis

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("./data/grades.csv", index_col = "Identifier")
```

```python
# This is the examples of existing grades of the images
df.head()
```

```python
# Plot histogram of final scores of cards
fig = px.histogram(df, x = 'Grade')
fig.show()
```

It can be seen that there is a very high number of trading cards whose grade is around 8.5 to 9.5. For other cards whose grade is lower than 8.5, the examples are limited, so it would be challenging for the model to learn low-quality trading cards.

```python
fig = px.histogram(df, x = 'Centering')
fig.show()
```

Grades for Centering are very bias, as it only concentrates at the range of 8.0 to 10.0.
