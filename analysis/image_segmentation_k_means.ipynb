```python
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
root_path = os.path.join('data','examples','images', 'cgc', '3727258001', 'front.jpg')
```

```python
image = cv2.imread(root_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
X = rgb_image.reshape((-1,3))
X = np.float32(X)
```

```python
# Using k-means algorithm with k = 1 to get the content
n_points, n_dim = X.shape
x_max, x_min = np.max(X), np.min(X)
centroids = np.random.uniform(low = x_min, high=x_max, size = (2,n_dim))
kmeans = KMeans(n_clusters = 2, init = centroids, max_iter = 1000).fit(X)
```

```python
cluster_0 = rgb_image.reshape((-1,3))
```

```python
cluster_0[np.where(kmeans.labels_ == 1),:] = 0
```

```python
cluster_0 = cluster_0.reshape((image.shape))
```

```python
plt.figure(figsize = (10,10))
plt.imshow(cluster_0)
```
